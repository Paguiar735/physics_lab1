```python
# Remover o comentário das duas linhas abaixo em caso seja usado compiladores mais avançados ou offline.
!pip install tabulate==0.8.7 --user
!pip install sympy==1.6.2 --user
!pip install scipy==1.5.2 --user

#########################################################
#--[Cálculo Numérico (Unidade 3) - Trabalho (Parte 1)]--#
#########################################################
#-----------------[Método dos Trapézios]----------------#
#########################################################

#
##
###
##########################################
# 1 - Instalação e Importação de Livrarias
##########################################
####
###
#

# Manipulação de Funções em Formato Simbólico (com o uso de variáveis)
import sympy as sp
import scipy as sci
from sympy.utilities.lambdify import lambdify, implemented_function
from scipy import optimize
from sympy import *

# Exibição de Tabelas ASCII
from tabulate import tabulate
 
#
##
###
###########################
# 2 - Definição das Funções
###########################
####
###
#

#
##
########################
# 2.0 - Funções Diversas
########################
##
#

def interface():
  horizontal_line(30, "Início do Script", 1, 2)
  border_msg('Integração pelo Método dos Trapézios v1.0')
  print("\n\u2022 Opções Disponíveis\n")
  print("1. Definição do Método (Digite 1)")
  print("2. Exibir resolução das Questões do Exercício (Digite 2)")
  print("3. Aplicar o Método em uma Função (Digite 3)\n")

  r = input("\u2022 Escolha uma opção: ")
  print("")
  if r == "1":
    print("\u2192 Opção Selecionada:", r, "\n\n")
    statement_of_rule()
  if r == "2":
    print("\u2192 Opção Selecionada:", r, "\n\n")
    answer_key()
  if r == "3":
    print("\u2192 Opção Selecionada:", r, "\n\n")
    application_of_method()
  if (r != "1") and (r != "2") and (r != "3"):
    print("\u2192 Opção Inválida")

def border_msg(msg):
    row = len(msg)+2
    h = ''.join(['+'] + ['-' *row] + ['+'])
    result= h + '\n'"| "+msg+" |"'\n' + h + "\n"
    print(result)

def horizontal_line(n, msg, u, b):
  n = int(n/2)
  print("\n"*u)
  print("-"*n + "[ " + msg + " ]" + "-"*n)
  print("\n"*b)

def local_maxima(func, a, b):
    g = -sp.diff(func, 'x', 2)
    g = lambdify('x', g)
    k = sci.optimize.minimize_scalar(g, bounds=[a,b], method='bounded')
    k = -g(k.x)
    return k
    
def error_bounds_f(f, func, a, b, n, p):
    k = local_maxima(func, a, b)
    E = (k/(12*n**2))*(b-a)**3
    print("\u2022 |E| =", round(E, p))

#
##
#####################################
# 2.2 - Opção 1 (Definição do Método)
#####################################
##
#

def statement_of_rule():
  border_msg(" 1 - Definição do Método ")
  print("\u2192 Seja f(x) uma função qualquer, temos que:\n")
  print("\u222B[a, b][f(x) dx] \u2248 f(x\u2081) + \u2211[i = 1, n-1][2f(x\u1D62)] + f(x\u2099) = T\u2099\n")
  print("\u2192 Ou seja:\n")
  print("T\u2099 = (h/2)[ f(x\u2080) + 2f(x\u2081) + 2f(x\u2082)+ ... + f(x\u2099) ]\n")
  print("\u2192 Onde:\n")
  print("\u2022 a = Limite Inferior\n\u2022 b = Limite Superior\n\u2022 n = Número de Subintervalos\n\u2022 h = (b-a)/h\n\u2022 k = Máximo Local\n\u2022 Erro: Erro = [k(b-a)^3]/(12n^2)")
  horizontal_line(30, "Fim do Script", 1, 1)
  
#
##
##########################################
# 2.2 - Opção 2 (Resolução das Atividades)
##########################################
##
#

def answer_key():
  p = 3
  #
  ### Questão 1-a
  #
  a, b, n = 1, 2, 4
  func = sp.exp(2*x)
  f = lambdify('x', func)
  border_msg(" 2.1 - Questão 1: Confirmação da Letra A (Q. 3 do Livro) ")
  print("\u2192 Função: f(x) =", func, "\n\u2192 Intervalo: [a, b] =", "[{}, {}]".format(str(a), str(b)), "\n\u2192 Número de Subintervalos: n =", n, "\n\u2192 Critério de Arredondamento: {} casas decimais".format(p), "\n")
  trapezoidal_rule(f, a, b, n, p)
  error_bounds_f(f, func, a, b, n, p)
  print("\n")
  #
  ### Questão 5-a
  #
  a, b = 0, 6
  list_intervals = [0, 1, 2, 3, 4, 5, 6]
  list_images = [0.21, 0.32, 0.42, 0.51, 0.82, 0.91, 1.12]
  border_msg(" 2.2 - Questão 5: Confirmação da Letra A (Q. 9 do Livro) ")
  print("\u2192 Função: f(x) = \u2205", "\n\u2192 Intervalo: [a, b] =", "[{}, {}]".format(str(a), str(b)), "\n\u2192 Número de Subintervalos: n =", n, "\n\u2192 Critério de Arredondamento: {} casas decimais".format(p), "\n")
  trapezoidal_rule_data_points(list_intervals, list_images, p)
  horizontal_line(30, "Fim do Script", 1, 1)

def trapezoidal_rule_data_points(list_intervals, list_images, p):
  h = float(list_intervals[1] - list_intervals[0])
  T = area(list_images, h, p)
  table(list_intervals, list_images)
  print("\n\u2022  T\u2099 =", T)

#
##
####################################################################
# 2.3 - Opção 3 (Integração com Funções e Intervalos Personalizados)
###################################################################
##
#

def application_of_method():
    p = 3
    func, a, b, n, p = user_input()
    f = lambdify('x', func)
    print("\n")
    border_msg(" 3 - Função e Intervalo Personalizados ")
    print("\u2192 Função: f(x) =", func, "\n\u2192 Intervalo: [a, b] =", "[{}, {}]".format(str(a), str(b)), "\n\u2192 Número de Subintervalos: n =", n, "\n\u2192 Critério de Arredondamento: {} casas decimais".format(p), "\n")
    trapezoidal_rule(f, a, b, n, p)
    horizontal_line(30, "Fim do Script", 1, 1)
    
def user_input():
  a = float(input("\u2192 Insira o Limite Inferior do Intervalo: a = "))
  b = float(input("\u2192 Insira o Limite Superior do Intervalo: b = "))
  n = int(input("\u2192 Insira o Número de Subintervalos: n = "))
  p = int(input("\u2192 Insira o Critério de Arredondamento (em Casas Decimais): p = "))
  func = str(input("\u2192 Insira a Lei da Função na variável x: f(x) = "))
  return func, a, b, n, p

def trapezoidal_rule(f, a, b, n, p):
  h = h_mean(a, b, n)
  list_intervals = intervals(a, h, n, p)
  list_images = images(list_intervals, f, p)
  T = area(list_images, h, p)
  table(list_intervals, list_images)
  print("\n\u2022  T\u2099 =", T)

def h_mean(a, b, n):
  h = (b-a)/n
  return h

def intervals(a, h, n, p):
  list_intervals=[]
  for i in range(0, n+1):    
    list_intervals.append(a)
    a = round(a + h, p)    
    i =+ 1
  return list_intervals
 
def images(list_intervals, f, p):
  list_images=[]
  for i in list_intervals:
    c = round(f(i), p)
    list_images.append(c)
  return list_images
 
def area(list_images, h, p):
  T = 0
  n = len(list_images) - 1 # Definição do "n" sem precisá-lo como parâmetro.
  for i in range(0, n+1):
    if (i == 0) or (i == n):
      T = list_images[i] + T
    else:
      T = 2*list_images[i] + T
  T = round((h/2)*T, p)
  return T

def table(list_intervals, list_images):
  list_table = []
  n = len(list_intervals) - 1 # Definição do "n" sem precisá-lo como parâmetro.
  for i in range(0, n+1):
    list_table.append([list_intervals[i], list_images[i]])
  print(tabulate(list_table, headers=['x', 'y'], tablefmt="github"))

#
##
###
########################
# 3 - Início do Programa
########################
####
###
#

x = sp.Symbol('x')
interface()

#########################################################
#-[Criado por Pedro Victor Aguiar da Silva em 12/09/20]-#
#########################################################
```
